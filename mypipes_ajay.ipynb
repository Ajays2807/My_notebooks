```python
class vaeselector(BaseEstimator,TransformerMixin):
    def __init__(self,features_names):
        
        self.features_names=features_names
        
    def fit(self,x,y=None):
        
        return self
    
    def transform (self,X):
        
        return X[self.features_names]
    
    def get_feature_names(self):
        
        return self.features_names
    
    
class get_dummies(BaseEstimator,TransformerMixin):
    
    def __init__(self,frequency_cutoff=0):
        
        self.features_names=[]
        self.frequency_cutoff=frequency_cutoff
        catg_dict={}
        
    def fit(self,x,y=None):
        
        data_cols=x.columns
        
        for col in data_cols:
            k=x[col].value_counts()
            if (k<=self.frequency_cutoff).sum()==0:
                
                cats=k.index[:-1]
                
            else:
                cats=k.index[k>self.frequency_cutoff]
                
            self.catg_dict[col]=cats 
            
        for col in catg_dict.keys():
            for cat in catg_dict[col]:
      
                self.features_names.append(col+'_'+str(cat))
        return self
    
    def transform(self,X):
        dummy_data=X.copy
        for col in catg_dict.keys():
            for cat in catg_dict[col]:
                names=col+'_'cat
                dummy_data[names]=(dummy_data[col]==cat).astype(int)
                
            del dummy_data[col]
        return dummy_data

    def get_feature_names(self):
        
        return self.features_names
                
        
        
```
