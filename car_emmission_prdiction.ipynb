```python
x=emmisions_data.drop('ppm',axis=1)
```

```python
x.head()
```

```python
import matplotlib.pyplot as plt
emmisions_data.plot(y=emmisions_data.columns,figsize=(10,10))
plt.xlabel('x axis')
plt.ylabel('parameters')
plt.show()
```

```python
y=emmisions_data['ppm']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.model_selection import cross_val_score
cvmae=-cross_val_score(lr,x_train,y_train,cv=10,n_jobs=-1,verbose=20,scoring='neg_mean_absolute_error')
```

```python
cvmae
```

```python
cvmae.mean()
```

```python
lr.fit(x_train,y_train)
```

```python
list(zip(lr.coef_,x_train.columns))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
pred_train=lr.predict(x_train)
score_train=mean_absolute_error(pred_train,y_train)
print(score_train)
```

```python
pred_test=lr.predict(x_test)
score_test=mean_absolute_error(pred_test,y_test)
print(score_test)
```

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV

```

```python
lamdas=np.linspace(100000,500000,500)
```

```python
params={'alpha':lamdas}
```

```python
rg=Ridge(fit_intercept=True)
gs=GridSearchCV(rg,param_grid=params,cv=10,scoring='neg_mean_absolute_error',verbose=20,n_jobs=-1)
```

```python
gs.fit(x_train,y_train)
```

```python
gs.best_estimator_
```

```python
ridge_fit=gs.best_estimator_
```

```python
ridge_fit.fit(x_train,y_train)
```

```python
pred=ridge_fit.predict(x_test)
```

```python
score_mae=mean_absolute_error(pred,y_test)
```

```python
score_mae
```

```python
ls=Lasso(fit_intercept=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
dt=DecisionTreeRegressor()
params={'max_depth':[5,10,15,20,30,50],
         'min_samples_split':[15,16,17,20],
          'min_samples_leaf': [1,2,5,10] }
```

```python
rs=RandomizedSearchCV(dt,
                      param_distributions=params,
                      cv=10,
                      verbose=20,
                      n_jobs=-1,
                      scoring='neg_mean_absolute_error')
```

```python
rs.fit(x_train,y_train)
```

```python
bestfit_Decisiontree= rs.best_estimator_
print(bestfit_Decisiontree)
```

```python
pred_train=bestfit_Decisiontree.predict(x_train)
pred_test=bestfit_Decisiontree.predict(x_test)
score_decisiontree_train=mean_absolute_error(pred_train,y_train)
score_decisiontree_test=mean_absolute_error(pred_test,y_test)
```

```python
score_decisiontree_test
```
